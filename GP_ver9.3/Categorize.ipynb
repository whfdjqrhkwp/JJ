```python
import import_ipynb

import Score_Noun
import Score_BoundNoun
import Score_Verb
import Score_Adverb
import Score_Interjection
import Score_Pronoun
import Score_Adjective
import Score_Affix
import Score_Determiner
import Score_Numeral

import Score_PreFinalEnding
import Score_FinalEnding
import Score_ConnectEnding
import Score_TransEnding
import Score_Postposition
```

```python
def categorize(pos):
    # 어휘, 문법 요소 및 형태소를 받음
    
    V = 'voca'
    G = 'gram'
    
    # 명사
    Noun = ['NN', 'NNG', 'NNP', 'UN']
    
    # 대명사
    Pronoun = ['NP']
    
    # 의존명사
    BoundNoun = ['NNB', 'NNM']
    
    # 동사
    Verb = ['VV', 'VXV']
    
    # 형용사
    Adjective = ['VA', 'VXA', 'VCN']
    
    # 수사
    Numeral = ['NR']
    
    # 부사
    Adverb = ['MA', 'MAC', 'MAG']
    
    # 접사
    Affix = ['XP', 'XPN', 'XPV', 'XSA', 'XSN', 'XSV']
    
    # 관형사
    Determiner = ['MD', 'MDN', 'MDT']
    
    # 감탄사
    Interjection = ['IC']
    
    # 조사
    Postposition = ['JC', 'JK', 'JKC', 'JKG', 'JKI', 'JKM', 'JKO', 'JKQ', 'JKS', 'JX', 'VCP']
    
    # 선어말 어미
    PreFinalEnding = ['EP', 'EPH', 'EPP', 'EPT']
    
    # 연결 어미
    ConnectEnding = ['EC', 'ECD', 'ECE', 'ECS', ]
    
    # 전성 어미
    TransEnding = ['ET', 'ETD', 'ETN']
    
    # 종결 어미
    FinalEnding = ['EF', 'EFA', 'EFI', 'EFN', 'EFO', 'EFQ', 'EFR']
    
    
    if pos[1] in Noun:
        return Score_Noun.scoring(pos[0]), V
    
    if pos[1] in Pronoun:
        return Score_Pronoun.scoring(pos[0]), V
    
    if pos[1] in BoundNoun:
        return Score_BoundNoun.scoring(pos[0]), V
    
    if pos[1] in Verb:
        return Score_Verb.scoring(pos[0]), V
    
    if pos[1] in Adjective:
        return Score_Adjective.scoring(pos[0]), V
    
    if pos[1] in Numeral:
        return Score_Numeral.scoring(pos[0]), V
        
    if pos[1] in Adverb:
        return Score_Adverb.scoring(pos[0]), V
    
    if pos[1] in Affix:
        return Score_Affix.scoring(pos[0]), V
    
    if pos[1] in Determiner:
        return Score_Determiner.scoring(pos[0]), V
    
    if pos[1] in Postposition:
        return Score_Postposition.scoring(pos[0]), G
    
    if pos[1] in PreFinalEnding:
        return Score_PreFinalEnding.scoring(pos[0]), G
    
    if pos[1] in ConnectEnding:
        return Score_ConnectEnding.scoring(pos[0]), G
    
    if pos[1] in TransEnding:
        return Score_TransEnding.scoring(pos[0]), G
    
    if pos[1] in FinalEnding:
        return Score_FinalEnding.scoring(pos[0]), G
        
    return 0, ''
```
